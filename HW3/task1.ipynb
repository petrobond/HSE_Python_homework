```python
import pandas as pd

df = pd.read_csv("Electronic_sales_Sep2023-Sep2024.csv")

df.head()
```

```python
for col in ["Total Price", "Add-on Total"]:
    df[col] = pd.to_numeric(df[col], errors="coerce")

df["Add-on Total"] = df["Add-on Total"].fillna(0)
```

```python
df_completed = df[df["Order Status"].str.lower().eq("completed")].copy()
```

```python
def most_frequent(s: pd.Series):
    vc = s.dropna().value_counts()
    return vc.idxmax() if not vc.empty else None

result = (
    df_completed
    .groupby("Customer ID")
    .agg(
        Preferred_Payment=("Payment Method", most_frequent),
        Total_Spent=("Total Price", "sum"),
        Addons_Spent=("Add-on Total", "sum"),
    )
    .round({"Total_Spent": 2, "Addons_Spent": 2})
    .sort_values("Total_Spent", ascending=False)
)

result.head()
```
